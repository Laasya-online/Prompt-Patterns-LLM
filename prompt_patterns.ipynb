2. Input Article

```python
article = """
Only one in every six million people have the Rh null blood type. Now researchers are trying to grow it in the laboratory in the hope it could save lives.

Blood transfusions have transformed modern medicine. If we are ever unlucky enough to be injured or need serious surgery, blood that has been donated by others can be life-saving.

But not everyone is able to benefit from this remarkable procedure. People with rare blood types struggle to find donated blood that will match their own.

One of the rarest – the Rh null blood type – is found in just 50 known people in the world. Should they ever be in an accident that needs a transfusion, their chances of getting one are slim. Those with Rh null are instead encouraged to freeze their own blood for long-term storage.

But, despite its rarity, this blood type is also highly prized for other reasons. Within the medical and research community it is sometimes referred to as "golden blood" due to how it can be used.

"""
article
```

3. Prompt Pattern 1 - Simple Summarizer

```python
prompt = f"""
Summarize the following article in 5 clear bullet points.

Article:
{article}
"""

response = client.chat_completion(
    messages=[
        {"role": "user", "content": prompt}
    ],
    max_tokens=250
)

print(response.choices[0].message["content"])
```

4. Prompt Pattern 2 - Executive Tone

```python
prompt_exec = f"""
Rewrite the following article in a concise, formal executive tone.
Imagine you are writing for a CEO.
Keep it under 200 words.

Article:
{article}
"""

response_exec = client.chat_completion(
    messages=[
        {"role": "user", "content": prompt_exec}
    ],
    max_tokens=300
)

print(response_exec.choices[0].message["content"])
```

5. Prompt Pattern 3 - Role + Constraints + Example

```python
prompt_pattern = f"""
You are a senior management consultant.

Task:
1. Summarize the article in exactly 4 bullet points.
2. Each bullet must be under 12 words.
3. Then write a short executive summary paragraph (max 120 words).

Example summary format:
- Key risk identified in current plan
- Growth stable across all segments
- Costs rising in two major regions
- Mitigation actions scheduled for next quarter

Now apply this style to the article below.

Article:
{article}
"""

response_pattern = client.chat_completion(
    messages=[
        {"role": "user", "content": prompt_pattern}
    ],
    max_tokens=400
)

print(response_pattern.choices[0].message["content"])
```

## 6. Comparison of Prompt Patterns

**Pattern 1 – Simple Summarizer**
- Directly asked for summary → clear, neutral tone
- No specific role or style → output is generic

**Pattern 2 – Executive Tone**
- Output became more formal, concise, and business-like
- Removed casual language and focused on key decisions/impact

**Pattern 3 – Role + Constraints + Example**
- Bullet points followed word limit more strictly
- Structure closely matched the example format
- Executive summary felt more structured and analytical
